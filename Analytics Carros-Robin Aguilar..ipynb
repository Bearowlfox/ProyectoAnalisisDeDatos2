```python
av_normalized = dfAutos["normalized"].astype("float").mean(axis=0)
av_bore = dfAutos["bore"].astype("float").mean(axis=0)
av_stroke = dfAutos["stroke"].astype("float").mean(axis=0)
av_horsepower = dfAutos["horsepower"].astype("float").mean(axis=0)
av_peak = dfAutos["peak"].astype("float").mean(axis=0)
```

```python
dfAutos["normalized"].replace(np.nan, 122, inplace = True)
dfAutos["bore"].replace(np.nan, 122, inplace = True)
dfAutos["stroke"].replace(np.nan, 122, inplace = True)
dfAutos["horsepower"].replace(np.nan, 122, inplace = True)
dfAutos["peak"].replace(np.nan, 122, inplace = True)
dfAutos.head(5)
```

<h5 Style = 'color:#8d3559'> 5.2 Comvertir Valores Numericos en Funcion del Medio (Estandarizar)</h5>


```python
dfAutos.head(4)
```

```python
dfAutos["City L/100km (Calc)"] = 235/dfAutos["City"]
dfAutos["Higway L/100km (Calc)"] = 235/dfAutos["Higway"]
dfAutos.head(5)
```

```python
dfAutos.dtypes
```

<h5 Style = 'color:#8d3559'> 5.3 Normalizar Valores entre 0 y 1 </h5>

```python
dfTemp = dfAutos[["City L/100km (Calc)","Higway L/100km (Calc)" ,"engineSize"]]
dfTemp.head(5)
```

```python
dfTemp.describe()
```

```python
dfAutos["City L/100km (norm)"] = dfAutos["City L/100km (Calc)"] /dfAutos["City L/100km (Calc)"].max()
dfAutos["Higway L/100km (norm)"] = dfAutos['Higway L/100km (Calc)'] /dfAutos['Higway L/100km (Calc)'].max()
dfAutos["engineSize (norm)"] = dfAutos['engineSize'] /dfAutos["engineSize"].max()

dfTemp = dfAutos[["City L/100km (Calc)","Higway L/100km (Calc)",'engineSize',"City L/100km (norm)",
                 "Higway L/100km (norm)","engineSize (norm)"]]
dfTemp.head(5)
```

<p>Pregunta 1.Â¿Cuantos autos existen segun marca?</p>


```python
gf=dfAutos.groupby(["make"]).count()
gf=gf.iloc[:,:1]
gf=gf.sort_values("symboling",ascending = False)
gf
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 6.9), dpi=100)

# AsegÃºrate de definir eje_x y data antes de utilizarlos
eje_x = gf.index.tolist()
data = gf["symboling"].tolist()

bar_colors = ['#17ffae', '#eddbed', '#eddbed', '#eddbed', '#eddbed', '#eddbed', '#eddbed', '#eddbed', '#eddbed', '#eddbed', '#eddbed', '#eddbed', '#eddbed', '#eddbed', '#eddbed', '#eddbed', '#eddbed', '#eddbed', '#eddbed', '#eddbed', '#eddbed']

# Cambia ax.barh para que las barras sean horizontales
ax.barh(eje_x, data, color=bar_colors)

# ConfiguraciÃ³n de tÃ­tulos
title = "Grafico 2: Vehiculos"
plt.suptitle(title,
             horizontalalignment='left',
             x=0.001,
             y=0.87,
             transform=fig.transFigure,
             fontsize=20,
             fontweight="bold",
             fontfamily="serif")

# Formato del subtitulo
subTitle = "Cantidad de vehiculos agrupados segun fabricante"
ax.set_title(subTitle,
             horizontalalignment='left',
             x=0.001,
             y=0.80,
             transform=fig.transFigure,
             fontsize=15,
             fontweight="regular",
             fontfamily="serif")

# Formato de los ejes
etiqueta = "Cantidad de vehiculos (ABS)"
ax.set_xlabel(etiqueta,
              ha='left',
              fontsize=10,
              fontweight="regular",
              fontfamily="serif")

# ConfiguraciÃ³n de coordenadas de la etiqueta del eje X
ax.xaxis.set_label_coords(x=0, y=-0.07, transform=ax.transAxes)

# ConfiguraciÃ³n del tamaÃ±o del texto del eje Y
ax.tick_params(axis='both',
               labelsize=10)

# ConfiguraciÃ³n de la ubicaciÃ³n del grÃ¡fico
plt.subplots_adjust(top=0.80)

plt.show()
```
